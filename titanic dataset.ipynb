```python
import pandas as pd
import os

X = pd.read_csv(r'C:\Users\abasit\Downloads\titanic_train.csv')
X.info()

y = X['survived']

X.pop("name")
X.pop("survived")
X.pop("home.dest")
X.pop("boat")



X = X.replace(to_replace ="male", value =0)
X = X.replace(to_replace ="female", value =1)

X.pop("ticket")
X.pop("cabin")
X.pop("embarked")
X.pop("sex")

X = X.fillna((X.mean()))




X




```

```python
print(X.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='lbfgs')

print(lr)
```

```python
lr.fit(X, y)
```

```python
X_test = pd.read_csv(r'C:\Users\abasit\Downloads\titanic_test.csv')


X_test.pop("name")


X_test = X_test.replace(to_replace ="male", value =0)
X_test = X_test.replace(to_replace ="female", value =1)
X_test.pop("ticket")
X_test.pop("cabin")
X_test.pop("embarked")
X_test.pop("home.dest")
X_test.pop("boat")
X_test.pop("sex")

X_test = X_test.fillna((X_test.mean()))

X_test


```

```python
y_pred = lr.predict(X_test)

type(X_test)
```

```python
submission = pd.DataFrame({"passenger_id": X_test["passenger_id"], "survived": y_pred})
```

```python
submission.to_csv('titanic.csv', index=False)
```
